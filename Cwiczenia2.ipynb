### NumPy

NumPy jest używane do pracy z tablicami i wykonywania operacji numerycznych.

#### Podstawowe operacje

````python
import numpy as np

# python array
pyarr = [1, 2, 3, 4, 5]
print(pyarr)

# numpy array
arr = np.array([1, 2, 3, 4, 5])
print(arr)

# Podstawowe operacje
print("Mediana:", np.mean(arr))
print("Odchylenie standardowe:", np.std(arr))
print("Suma:", np.sum(arr))
````

#### Wektory i macierze

````python
# Tworzenie tablic NumPy (1D i 2D)
a = np.array([1, 2, 3])            # wektor
b = np.array([[1, 2], [3, 4]])     # macierz 2x2

# Operacje matematyczne
c = a + 10             # dodaj 10 do każdego elementu
d = b * 2              # mnożenie przez skalar
e = a * a              # mnożenie elementów (mnożenie składnikowe)
f = np.dot(a, a)       # iloczyn skalarny

# Operacje na macierzach
dot = np.dot(b, b)       # mnożenie macierzy
transpose = np.transpose(b)    # transpozycja macierzy
inverse = np.linalg.inv([[1, 2], [3, 4]])  # macierz odwrotna (jeśli jest odwracalna)

print("Macierz b * b:\n", dot, "\n")
print("Transpozycja macierzy b:\n", transpose, "\n")
print("Macierz odwrotna macierzy [[1, 2], [3, 4]]:\n", inverse, "\n")
````

#### Slicing

````python
# 1d
arr = np.arange(1, 10)
a = arr[2:4]     # Elementy od indeksu 2 do 4
b = arr[::2]     # Co drugi element
c = arr[-2:]     # Ostatnie 2 elementy
d = arr[::-2]    # Co drugi element w odwrotnej kolejności

print("a) ", a)  # [3 4]
print("b) ", b)  # [1 3 5 7 9]
print("c) ", c)  # [8 9]
print("d) ", d)  # [9 7 5 3 1]

print()

# 2d
arr2d = np.arange(20).reshape(4, 5)  # przekształcenie (wiersze, kolumny)
print("Oryginalna tablica:\n", arr2d)

a2d = arr2d[:3, -3:]  # górny prawy róg 3x3
b2d = arr2d[2]        # 3-ci wiersz

print()
print(a2d)
print(b2d)
````

#### Porównanie wydajności - Pętla for, a operacje na wektorach





````python
import numpy as np
import time

n = 1_000_000
a = np.arange(n)
b = np.arange(n)

# Używanie pętli for
start_loop = time.time()
c_loop = []
for i in range(n):
    c_loop.append(a[i] + b[i])
end_loop = time.time()

# Używanie operacji na wektorach z pomocą NumPy
start_vector = time.time()
c_vector = a + b
end_vector = time.time()

print(f"Czas pętli for: {end_loop - start_loop:.5f} sekundy")
print(f"Czas operacji na wektorach: {end_vector - start_vector:.5f} sekundy")
````

Przydatne parametry

`skiprows=1`:
- To polecenie wskazuje bibliotece Pandas, aby **pominęło pierwszy wiersz** podczas wczytywania pliku CSV.
- Zazwyczaj `skiprows` jest używane, gdy pierwszy wiersz zawiera metadane, dodatkowe informacje lub dane nienależące do tabeli (np. nagłówki w niestandardowej pozycji lub komentarze).
- W takim przypadku pierwszy wiersz nie zostanie uwzględniony w DataFrame. Na przykład, jeśli plik zawiera wiersz wprowadzający lub jakąś opisową treść na górze pliku, ten parametr zapewnia, że będzie to zignorowane podczas wczytywania rzeczywistych danych.

`header=None`:
- Domyślnie `read_csv()` zakłada, że pierwszy wiersz zawiera **nagłówki kolumn** (nazwy kolumn).
- Ustawiając `header=None`, wskazuje to bibliotece Pandas, aby **nie traktowało żadnego wiersza jako nagłówki kolumn**. Zamiast tego Pandas automatycznie przypisze numeryczne etykiety kolumn (0, 1, 2, itd.).

1.Mediana to wartość środkowa w uporządkowanym zbiorze danych. Oznacza, że połowa obserwacji jest mniejsza, a połowa większa od mediany.

Jak ją znaleźć:

1.Uporządkuj dane rosnąco.
2a. Jeśli liczba obserwacji jest nieparzysta, mediana to środkowy element.
2b. Jeśli liczba obserwacji jest parzysta, mediana to średnia arytmetyczna dwóch środkowych wartości.
Przykłąd :
Przykład:
Zbiór: 3, 5, 7, 8, 12
- Liczba elementów = 5 (nieparzysta)
- Mediana = 7

Zbiór: 2, 4, 6, 8
- Liczba elementów = 4 (parzysta)
- Mediana = (4 + 6) / 2 = 5.




2. Średnia arytmtyczna - to suma wszystkich wartości podzielona przez ich liczbę.
Liczby: 2,4,6,10

średnia : (4+6+8+10)/4 = 7

3. Odchylenie standardowe mierzy, jak bardzo dane różnią się od średniej. Pokazuje, czy wartości są „rozrzucone” (duże odchylenie) czy „zbite” wokół średniej (małe odchylenie).

 Dane: 2,4,6
 1. liczmy średnią= 4
 2. liczymy wariancje: (
  (2 − 4)² + (4 − 4)² + (6 − 4)²) / 3 =
  (4 + 0 + 4) / 3 = 8 / 3 ≈ 2,67
  3. Liczymy odchylenie standardowe:
    √2,67 ≈ 1,63


4. Kwartyle-  to wartości, które dzielą uporządkowany zbiór danych na cztery równe części.
Każdy kwartyl pokazuje, w którym miejscu znajduje się określony procent danych:

Pierwszy kwartyl (Q1) – to wartość, poniżej której znajduje się 25% wszystkich danych.

Trzeci kwartyl (Q3) – to wartość, poniżej której znajduje się 75% wszystkich danych.


Jak obliczyć kwartyle:

- Uporządkuj dane od najmniejszej do największej wartości.

- Znajdź medianę całego zbioru (to drugi kwartyl – Q2).

- Q1 to mediana dolnej połowy danych (czyli części przed medianą).

- Q3 to mediana górnej połowy danych (czyli części po medianie).

Przykład:
Zbiór danych: [2, 4, 6, 8, 10, 12, 14]

Dane są już uporządkowane.

Mediana (środkowa wartość) = 8

Dolna połowa danych: [2, 4, 6] → Q1 = 4

Górna połowa danych: [10, 12, 14] → Q3 = 12



5. Rozstęp międzykwartylowy - to miara rozproszenia danych, która pokazuje, jak szeroko rozłożone są środkowe 50% wartości w zbiorze danych.
Innymi słowy – IQR mówi, jak duży jest odstęp między pierwszym (Q1) a trzecim (Q3) kwartylem.

Im większy IQR, tym bardziej zróżnicowane są dane w środkowej części zbioru.

Jak się go oblicza:

- Oblicz pierwszy kwartyl (Q1).

- Oblicz trzeci kwartyl (Q3).

- Odejmij Q1 od Q3 według wzoru:
#    IQR = Q3 − Q1

Przykład:
Zbiór danych: [2, 4, 6, 8, 10, 12, 14]
Z poprzednich obliczeń:

Q1 = 4

Q3 = 12

Obliczamy:
IQR = Q3 − Q1 = 12 − 4 = 8

6. Wariancja - to miara rozproszenia danych – pokazuje, jak bardzo wartości w zbiorze danych różnią się od średniej.

Jeśli wariancja jest duża, dane są bardzo rozproszone.

Jeśli wariancja jest mała, dane są blisko średniej.

Związek z odchyleniem standardowym - odchylenie standardowe to pierwiastek kwadratowy z wariancji

Dane: 2,4,6,8,10
- średnia:6
- odchylenie od średniej -4,-2,0,2,4 podnośmyu je do kwadratu
- wariancja (16+4+0+4+16)/5 = 8
